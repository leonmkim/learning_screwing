```python
import os
import argparse

import cv2

import rosbag
from sensor_msgs.msg import Image
from cv_bridge import CvBridge

import numpy as np
```

```python
bag_path = '~/datasets/contact/contact_dset_test_2022-07-13-14-17-57.bag'
bag_path = os.path.expanduser(bag_path)
os.path.exists(bag_path)
```

```python
bag_path_split = bag_path.split('/')
bag_path_root = '/'.join(bag_path_split[0:-1]) + '/'
bag_path_root
```

```python
bag_name = bag_path_split[-1][0:-4]
```

```python
if not os.path.exists(bag_path_root + bag_name):
    os.mkdir(bag_path_root + bag_name)
# output_dir = 
```

```python
bag = rosbag.Bag(bag_path, "r")
```

```python
topics = list(bag.get_type_and_topic_info()[1].keys())
topics
```

```python
topic_types = list(bag.get_type_and_topic_info()[1].values())
topic_types
```

```python
image_topics = []
im_substring = 'camera'
pcl_substring = 'points'
# im_topic = '/camera/color/image_raw/compressed'
for topic in topics:
    if im_substring in topic:
        if pcl_substring not in topic:
            image_topics.append(topic) 
image_topics
```

```python
image_topics[0].split('/')[2]
```

```python
bridge = CvBridge()
```

```python
for im_topic in image_topics:
    im_type = im_topic.split('/')[2]
    if not os.path.exists(bag_path_root + bag_name + '/' + im_type):
        os.mkdir(bag_path_root + bag_name + '/' + im_type)
    count = 0
    time_np = []
    for topic, msg, t in bag.read_messages(topics=im_topic):
        # cv_img = bridge.imgmsg_to_cv2(msg, desired_encoding="passthrough")
        cv_img = bridge.compressed_imgmsg_to_cv2(msg, desired_encoding="passthrough")

        # print(msg.header.stamp)
        time_np.append(t.to_time())
        # print(msg.header.seq)
        # print("{0}-{1}.png".format(count, msg.header.stamp))
        cv2.imwrite(os.path.join(bag_path_root + bag_name + '/' + im_type + '/', "{0}-{1}.png".format(count, t)), cv_img)
        print("Wrote image %i" % count)

        count += 1
        # if count == 2:
            # break
    np.save(bag_path_root + bag_name + '/' + im_type + '/' + 'timestamps', np.array(time_np))
```
